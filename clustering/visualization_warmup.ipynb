## Visualization Warmup

- Using mgp dataset

```python
from pydataset import data
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Acquire

```python
df= data('mpg')
```

## Create Grid

```python
plt.rc('figure', figsize=(12, 9))
plt.rc('axes', grid=True)
plt.rc('axes.spines', top=False, right=False, left=True, bottom=True)
plt.rc('font', size=14)
```

## Scatter Plot

- Hwy (x) Vs. Displ (y)

- Add meaningful labels and title

- Add horizontal dotted line that indicates the avg hwy mileage

- Add a vertical dotted line that indicates average engine displ

- Use color to indicate the number of cyl

- Instead of color, use separate subplots to indicate the number of cyl

```python
plt.scatter(df.displ, df.hwy)
plt.xlabel('Engine Displacement (L)')
plt.ylabel('Highway Mileage (MPG)')
plt.title('How does engine displacement relate to highway mileage?')

plt.hlines(df.hwy.mean(), df.displ.min(), df.displ.max(), ls=':', color='red')
plt.vlines(df.displ.mean(), df.hwy.min(), df.hwy.max(), ls=':', color='red')
plt.show()
```

## Relplot

```python
sns.relplot(data=df, x='displ', y='hwy', hue='cyl')
```

## Loop Scatter Plot

```python
for cyl in sorted(df.cyl.unique()):
    subset=df[df.cyl == cyl]
    plt.scatter(subset.displ, subset.hwy, label=cyl)
    
plt.xlabel('Engine Displacement (L)')
plt.ylabel('Highway Mileage (mpg)')
plt.title('How does engine displacement relate to highway mileage?')

plt.hlines(df.hwy.mean(), df.displ.min(), df.displ.max(), ls=':', color='firebrick', label='Highway Mileage Average')
plt.vlines(df.displ.mean(), df.hwy.min(), df.hwy.max(), ls=':', color='navy', label='Engine Displacement Average')
plt.show()
```

### Add Annotations

```python
for cyl in sorted(df.cyl.unique()):
    subset = df[df.cyl == cyl]
    plt.scatter(subset.displ, subset.hwy, label=cyl)
plt.legend(title='# Cylinders')
    
plt.xlabel('Engine Displacement (L)')
plt.ylabel('Highway Mileage (df)')
plt.title('How does engine displacement relate to highway mileage?')

plt.hlines(df.hwy.mean(), df.displ.min(), df.displ.max(), ls=':', color='red')
plt.vlines(df.displ.mean(), df.hwy.min(), df.hwy.max(), ls=':', color='red')

plt.annotate(
    f'Avg Engine Displacement: {df.displ.mean():.2f}',
    xy=(df.displ.mean(), 40),
    xytext=(3.7, 38),
    arrowprops={'arrowstyle': '->'}
)

plt.annotate(
    f'Avg Highway Mileage: {df.hwy.mean():.2f}',
    xy=(5, df.hwy.mean()),
    xytext=(5, 32),
    arrowprops={'arrowstyle': '->'}
)

plt.show()
```

### Create Subplots

```python
for i, cyl in enumerate(df.cyl.unique()):
    plt.subplot(2, 2, i + 1)
    subset = df[df.cyl == cyl]
    plt.scatter(subset.displ, subset.hwy, label=cyl)
    
    plt.title(f'{cyl} cylinders', size=12)
    plt.xlabel('Engine Displacement (L)', size=12)
    plt.ylabel('Highway Mileage (df)', size=12)
    
    plt.hlines(df.hwy.mean(), df.displ.min(), df.displ.max(), ls=':', color='red')
    plt.vlines(df.displ.mean(), df.hwy.min(), df.hwy.max(), ls=':', color='red')

plt.subplots_adjust(hspace=.3)
plt.suptitle('How does engine displacement relate to highway mileage?')
```
