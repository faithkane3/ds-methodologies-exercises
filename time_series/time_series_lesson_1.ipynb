## asfreq()

- .ffill()

- .bfill()

- fillna(0)

```python
# this df is missing weekend date, the dates are not distanced uniformly
# the outcome though is NaNs bc the created rows have no values; you will need to impute

df.asfreq('D')
```

```python
# forward fill looks like this, but it doesn't make sense here bc coffee consumption on weekends will be different

df.asfreq('D').ffill()
```

```python
# backward fill

df.asfreq('D').bfill()
```

```python
# fill with 0

df.asfreq('D').fillna(0)
```

## .resample

- like groupby, so it must have an aggregate function with it

```python
# this is like a groupy, so it must have an aggregate function with it or it just creates an object

df.resample('M').sum()

df.resample('M').agg(['min', 'mean', 'max', 'count'])
```

```python
df.resample('W').agg(['mean', 'count'])
df.resample('3W').agg(['mean', 'count'])
```

```python
# rolling or moving average

jan = df['2019'].copy()
jan = df.asfreq('D').fillna(0)
jan['3-day-avg'] = jan.rolling(3).mean()
```

```python
# first two days have NaNs bc they don't have a three day avg, only two days

jan
```

## .shift()

```python
# lag and lead operations; you can shift your data forward a day, back a day, forward or back n_days

feb = df['2019-02'].copy()
feb = feb.asfreq('D').fillna(0)
feb['shift-1'] = feb.shift(1)
feb['shift-3'] = feb.coffee_consumption.shift(3)
```

# Temps Exercises

```python
temps = data.sf_temps()
```

```python
temps.info()
```

```python
temps.head()
```

## Resample by the day and take the average temperature. Visualize the average temperature over time.


```python
temps = temps.set_index('date')
```

```python
temps.head()
```

```python
temps.asfreq('D')
```

```python
temps.resample('D').mean()
```

```python
temps.plot()
plt.title('Average Temps')
plt.ylabel('Degrees F')
plt.show()
```

## Write the code necessary to visualize the minimum temperature over time.


```python
temps.resample('M').min().plot()
plt.title('Minimum Temps by Month')
plt.ylabel('Degrees F')
plt.show()
```

## Write the code necessary to visualize the maximum temperature over time.


```python
temps.resample('M').max().plot()
plt.title('Maximum Temps by Month')
plt.ylabel('Degrees F')
plt.show()
```

## Which month is the coldest, on average?


```python
temps.resample('M').mean().sort_values(by='temp').head(1)
```

## Which month is the warmest, on average?

```python
temps.resample('M').mean().sort_values(by='temp', ascending=False).head(1)
```

## Resample by the day and calculate the min and max temp for the day (Hint: .agg(['min', 'max'])). Use this resampled dataframe to calculate the change in temperature for the day. Which month has the highest daily temperature variability?


```python
min_max = temps.temp.resample('D').agg(['min', 'max'])
```

```python
# min_max['temp_diff'] =(min_max.temp['max']) - (min_max.temp['min'])

min_max['diffs'] = min_max['max'] - min_max['min']
min_max['diffs'].plot()
plt.show()
```

#### highest daily temperature variability by month

```python
min_max.resample('M').mean().sort_values(by='diffs', ascending=False).head(1)
```

## Bonus: Visualize the daily min, average, and max temperature over time on a single line plot, i.e. the min, average, and maximum temperature should be 3 seperate lines.

```python
# viz of greatest change in daily temp by month: August

min_max.asfreq('M').plot()
plt.title('Temps by Month')
plt.ylabel('Degrees F')
plt.legend()
plt.show()
```

# Seatle Weather Exercises

```python
weather = data.seattle_weather()
```

```python
weather.info()
```

```python
weather.head()
```

## Which year and month combination has the highest amount of precipitation?


```python
weather = weather.set_index('date')
weather.head()
```

#### Month and year with the highest amount of precipitation

```python
weather.resample('M').max().sort_values(by='precipitation', ascending=False).head(1)
```

## Visualize the amount of monthly precipitation over time.


```python
weather.precipitation.resample('M').sum().plot()
plt.title('Amount of Precipitation by Month')
plt.ylabel('Inches of Precip')
plt.show()
```

## Visualize the amount of wind over time. Choose a time interval you think is appropriate.


```python
weather.wind.resample('D').sum().plot()
plt.title('Amount of Wind by Day')
plt.ylabel('Wind Units')
plt.show()
```

## Which year-month combination is the windiest?


```python
weather.resample('M').max().sort_values(by='wind' , ascending=False).head(1)
```
